**Which month had the highest number of cancelations?**

```python
resort_bookings_per_month = data.loc[data['hotel'] == 'Resort Hotel'].groupby('arrival_date_month')['hotel'].count()
resort_cancelations_per_month = data.loc[(data['hotel'] == 'Resort Hotel')].groupby('arrival_date_month')['is_canceled'].sum()

city_bookings_per_month = data.loc[data['hotel'] == 'City Hotel'].groupby('arrival_date_month')['hotel'].count()
city_cancelations_per_month = data.loc[(data['hotel'] == 'City Hotel')].groupby('arrival_date_month')['is_canceled'].sum()


resort_cancelations_data = pd.DataFrame({'Hotel' : 'Resort Hotel',
                                         'Month':list(resort_bookings_per_month.index),
                                         'Bookings': list(resort_bookings_per_month.values),
                                         'Cancelations': list(resort_cancelations_per_month.values)})


city_cancelations_data = pd.DataFrame({'Hotel' : 'City Hotel',
                                         'Month':list(city_bookings_per_month.index),
                                         'Bookings': list(city_bookings_per_month.values),
                                         'Cancelations': list(city_cancelations_per_month.values)})

cancelations_data = pd.concat([resort_cancelations_data, city_cancelations_data], ignore_index = True)
cancelations_data


```

```python
trace = go.Bar(x = city_cancelations_data['Month'],
                y = city_cancelations_data['Cancelations'],
                name = 'Resort Canceled'
)

trace1 = go.Bar(x = resort_cancelations_data['Month'],
                y = resort_cancelations_data['Cancelations'],
                name = 'City Canceled'
)

data6 = [trace, trace1]
layout = go.Layout(
    title = 'Cancelations by Month'
)

fig = go.Figure(data=data6, layout=layout)
pyo.plot(fig)
```
